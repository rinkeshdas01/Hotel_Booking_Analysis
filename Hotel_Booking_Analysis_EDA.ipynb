#### Room mismatch % for hotels

````python
# Chart - 7 visualization code
# We determine the room mismatch percentage
list_rr=df1['reserved_room_type'].tolist()
list_ar=df1['assigned_room_type'].tolist()
room_mismatch=[]
for i,j in zip(list_rr,list_ar):
  if i==j:
    room_mismatch.append(0)
  else:
    room_mismatch.append(1)
#Determining room mismatching % for hotels
df1['room_mismatch']=room_mismatch
df_room_mm=df1.groupby(['hotel'],as_index=False).agg({'room_mismatch':['sum','count']})
df_room_mm.columns=['Hotel','Rooms mismatched','No. of bookings']
df_room_mm['Room mismatch %']=(df_room_mm['Rooms mismatched']/df_room_mm['No. of bookings'])*100
df_room_mm=df_room_mm.drop(['Rooms mismatched','No. of bookings'],axis=1)
df_room_mm
#plotting a graph of room mismatch % for the hotels
plt. rcParams['figure.figsize']=8,8
plot_rmm=sns.barplot(x='Hotel',y='Room mismatch %',data=df_room_mm,palette="husl")
plt.xlabel('Hotel',fontsize=14)
plt.ylabel('Room mismatch %', fontsize=14)
plt.title('Room mismatch % for Hotels', fontsize=16)
for bar in plot_rmm.patches:
  plot_rmm.annotate(format(bar.get_height(),'.2f'),(bar.get_x()+bar.get_width()/2,bar.get_height()),ha='center',va='center',size=13,xytext=(0,9),textcoords='offset points')
````

##### 1. Why did you pick the specific chart?

Rooms mismatch happens when a guest is allotted a different room type from the room type of their choice. This leads to dissatisfaction of the guest.

##### 2. What is/are the insight(s) found from the chart?

The chart suggests that Resort Hotel has a higher room mismatch % as compared to City Hotel.

##### 3. Will the gained insights help creating a positive business impact? 
Are there any insights that lead to negative growth? Justify with specific reason.

According to the insights the guests of Resort Hotel might be disappointed by the high room mismatch %. This leads to loss of trust of the customer in the hotel and might lead to less bookings in the future.
